```julia
using SymPy

# Define parameters
E_0 = 1
a = 1
k = 1
z = 0
x0 = a/2
```

```julia
function V(x)
    return E_0/2 * (1 - cos(2π*x/a)) + k/2 * (z - x)^2
end
```

```julia
function V_taylor(x)
    return V(x0) + (x - x0)*(E_0*π/a * sin(2*π*x0/a) - k*(z - x0)) + (x - x0)^2 * ((2*π*E_0/a)^2 * cos(2*π*x0/a) + k)/2 + (x - x0)^3 * (-(2*π*E_0/a)^3 * sin(2*π*x0/a))/6
    end
```

```julia
x = range(-5, 5, length=1000)
```

```julia
using Plots

plot(x, V.(x), label="Potencial Tomlinson", xlims=(-0.25,1.25), ylims=(-0.25,1.25),lw=4)
plot!(x, V_taylor.(x), label="Aproximación Taylor", lw=4)
xlabel!("x")
ylabel!("V(x)")
```
