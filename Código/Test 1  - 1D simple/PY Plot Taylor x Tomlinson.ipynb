```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
import math

# Define parameters
E_0 = 1
a = 1
k = 1
z = 0
x0 = a/2
```

```python

# Define the Tomlinson potential
def V(x):
    return E_0/2 * (1 - np.cos(2*np.pi*x/a)) + k/2 * (z - x)**2

# Define the Taylor approximation
def V_taylor(x):
    return V(x0) + (x - x0)*(E_0*np.pi/a*np.sin(2*np.pi*x0/a) - k*(z - x0)) + (x - x0)**2*((2*np.pi*E_0/a)**2*np.cos(2*np.pi*x0/a) + k)/2 + (x - x0)**3*(-(2*np.pi*E_0/a)**3*np.sin(2*np.pi*x0/a))/6

# Generate x values
x = np.linspace(-5, 5, 1000)
```

```python

# Plot the Tomlinson potential and its Taylor approximation
plt.plot(x, V(x), label='Potencial Tomlinson')
plt.plot(x, V_taylor(x), label='Aproximación Taylor')
plt.xlabel('x')
plt.ylabel('V(x)')
plt.legend()
plt.ylim(-0.5,1.5)
plt.xlim(-0.5,2)
plt.show()
```
